## Extract response from Pushover Analysis (1-story building)

```python
# This file is used to extract the data from OpenSees pushover analysis result
# and generate pushover curves
# Developed by GUAN, XINGQUAN @ UCLA April 2019

# Import necessary modules
import numpy as np
import os
import matplotlib.pyplot as plt
import pickle

# Base directory where all the building models are stored
base_directory = 'C:\\Users\\61946\\Documents\\GitHub\\SeismicDesignModule\\BuildingNonlinearModels'

# The directory where all prerequisite python files exists
root_directory = 'C:\\Users\\61946\\Documents\\GitHub\\SeismicDesignModule'

# Analysis type: Pushover, EigenValueAnalysis, DynamicAnalysis
analysis_type = 'PushoverAnalysis'

# Define the output folder
if analysis_type == 'PushoverAnalysis':
    output_folder = 'PushoverOutput'
elif analysis_type == 'EigenValueAnalysis':
    output_folder = 'EigenAnalysisOutput'
elif analysis_type == 'DynamicAnalysis':
    output_folder = 'IDAOutput'
else:
    print('Wrong analysis type specified!')

# Building data folder where the building design results are stored
data_directory = 'C:\\Users\\61946\\Documents\\GitHub\\SeismicDesignModule\\BuildingData'
    
# Import all predefiend packages
os.chdir(root_directory)

from global_variables import SECTION_DATABASE
from global_variables import COLUMN_DATABASE
from global_variables import BEAM_DATABASE
from model_generation import model_generation
from building_information import Building
```

```python
# Loop over each folder and extract the results
all_drifts = []
all_reactions = []
all_weights = []
IDs = range(81, 567, 3)
for id in IDs:
    building_id = 'Building_' + str(id)
    # The folder path where reactions are stored
    reaction_directory = base_directory + '\\' + building_id + '\\' + analysis_type + '\\' + output_folder + '\\BaseReactions'
    # The folder path where roof drifts are stored
    drift_directory = base_directory + '\\' + building_id + '\\' + analysis_type + '\\' + output_folder + '\\StoryDrifts'
    # The folder where design data is stored
    design_directory = data_directory + '\\' + building_id
    # Load the reaction
    os.chdir(reaction_directory)
    reaction = np.sum(np.loadtxt('XReactions.out')[:, 1:], 1)
    # Load the roof drifts
    os.chdir(drift_directory)
    drift = np.loadtxt('Roof.out')[:, -1]
    # Load the design results for seismic weight
    os.chdir(design_directory)
    with open('construction_building.pkl', 'rb') as file:
        building = pickle.load(file)
    weight = sum(building.gravity_loads['floor weight'])
    # Attach each building's result to the list for whole building database
    all_drifts.append(drift)
    all_reactions.append(reaction)
    all_weights.append(weight)
```

```python
# Define parameters for the figure
FIGSIZE = (8, 8*0.618)
FONTSIZE = 16
# Plot the pushover curves
fig = plt.figure(figsize = FIGSIZE)
for indx in range(len(IDs)):
    plt.plot(all_drifts[indx], -all_reactions[indx]/all_weights[indx], color=(139/255, 139/255, 120/255))
plt.ylim([0, 1.0])
plt.xlim([0, 0.10])
plt.title('Pushover results for 1-story building', fontname='Times New Roman', fontsize=FONTSIZE, fontweight='light')
plt.xlabel('Story drift', fontname='Times New Roman', fontsize=FONTSIZE, fontweight='light')
plt.ylabel('Normalized reaction', fontname='Times New Roman', fontsize=FONTSIZE, fontweight='light')
plt.xticks(fontname='Times New Roman', fontsize=FONTSIZE, fontweight='light')
plt.yticks(fontname='Times New Roman', fontsize=FONTSIZE, fontweight='light')
plt.show()
```
